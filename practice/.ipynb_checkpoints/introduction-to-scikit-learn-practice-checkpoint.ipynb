## 1. Getting our ready to be used with machine learning

Three main things we have to do:
    
1. Split the data to features and labels (usually `X` and `y`)
2. Filling (also called imputing) or disregarding missing values
3. Coverting non-numberical values to numberical values (also called feature encoding)


```python
# Import our data and view it
heart_disease = pd.read_csv("../data/heart-disease.csv")
heart_disease.head()
```

```python
X = heart_disease.drop("target", axis=1)
X.head()
```

```python
y = heart_disease["target"]
y.head()
```

```python
# Split the data into training and test sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.2)
```

### 1.1 Make sure it's all numerical

```python
car_sales = pd.read_csv("../data/car-sales-extended.csv")

car_sales.head()
```

```python
len(car_sales)
```

```python
car_sales.dtypes
```

```python
# Set up random seed
np.random.seed(42)

# Create X and y
X = car_sales.drop("Price", axis=1)
y = car_sales["Price"]

# Split the data to training and test set
from sklearn.model_selection import train_test_split

X_tarin, X_test, y_tain, y_test = train_test_split(X,
                                                   y,
                                                   test_size=0.2)
```

```python
# Trun the categories into numbers

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ["Make", "Colour", "Doors"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",
                                   one_hot,
                                   categorical_features)],
                                remainder="passthrough")

transformed_X = transformer.fit_transform(X)
transformed_X
```

```python
# Another way to change numerical
dummies = pd.get_dummies(car_sales[["Make", "Colour", "Doors"]])
dummies
```

```python
# Let's refit the model

from sklearn.ensemble import RandomForestRegressor

np.random.seed(42)

X_train, X_test, y_train, y_test = train_test_split(transformed_X,
                                                    y,
                                                    test_size=0.2)

model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

### 1.2 What if there were missing values?

```python
# Import car sales missing data
car_sales_missing = pd.read_csv("../data/car-sales-extended-missing-data.csv")

car_sales_missing.head()
```

```python
# Create X and y

X = car_sales_missing.drop("Price", axis=1)
y = car_sales_missing.Price
```

```python
# Let's try and convert our data to numbers
# Trun categories into nubers

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ["Make", "Colour", "Doors"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",
                                   one_hot,
                                   categorical_features)],
                                remainder="passthrough")

transformed_X = transformer.fit_transform(X)
transformed_X
```

### Option 1: Fill the missing data with Pandas

```python
# Fill the "Make" column
car_sales_missing["Make"].fillna("Toyota", inplace=True)

# Fill the "Colour" column
car_sales_missing["Colour"].fillna("White", inplace=True)

# Fill the "Doors" column
car_sales_missing["Doors"].fillna(4, inplace=True)

# Fill the "Odometer (KM)" column
car_sales_missing["Odometer (KM)"].fillna(car_sales_missing["Odometer (KM)"].mean(),
                                          inplace=True)
```

```python
car_sales_missing.isna().sum()
```

```python
# Remove rows with missing Price value
car_sales_missing.dropna(inplace=True)
```
